```python
from blind_watermark import WaterMark
from tqdm import tqdm
import os
import re
import cv2
from imwatermark import WatermarkEncoder
```

```python


def str2msg(str):
    return [True if el=='1' else False for el in str]
key = '111010110101000001010111010011010100010000100111' # model key
bool_key = str2msg(key)
bwm1 = WaterMark(password_img=1, password_wm=1)



def batch_process_images(input_dir, output_dir, pattern=r"[wd]_\d+\.png")
    """
    批量处理图像
    
    参数:
    input_dir: 输入图像目录
    output_dir: 输出图像目录
    pattern: 文件名匹配模式
    """
    # 确保输出目录存在
    os.makedirs(output_dir, exist_ok=True)
    
    # 获取所有图像文件
    all_files = os.listdir(input_dir)
    # 筛选符合模式的文件
    pattern = re.compile(pattern)
    image_files = [f for f in all_files if pattern.match(f)]
    
    print(f"找到 {len(image_files)} 个文件符合条件")
    
    # 处理每个图像
    for img_file in tqdm(image_files, desc="处理图像"):
        input_path = os.path.join(input_dir, img_file)
        output_path = os.path.join(output_dir, img_file)
        
        bwm1.read_img(input_path)
        bwm1.read_wm(bool_key, mode='bit')
        bwm1.embed(output_path)


if __name__ == "__main__":
    # 输入和输出目录
    input_dir = "/home/jie/Program/SG_images/val/d"
    output_dir = "/home/jie/Program/blind_watermark/output/DCTDWT_Watermark"
    
    # 批量处理图像
    batch_process_images(input_dir, output_dir, pattern=r"d_\d+\.png")

    
    print("批量处理完成！")
    
```

```python
def Bit_acc(input_dir, pattern=r"[wd]_\d+\.png"):
    All_bit_acc = 0
    
    bwm1 = WaterMark(password_img = 1, password_wm = 1)

    all_files = os.listdir(input_dir)
    # 筛选符合模式的文件
    pattern = re.compile(pattern)
    image_files = [f for f in all_files if pattern.match(f)]
    
    print(f"找到 {len(image_files)} 个文件符合条件")
    
    # 处理每个图像
    for img_file in tqdm(image_files, desc="处理图像"):
        input_path = os.path.join(input_dir, img_file)
        wm_extract = bwm1.extract(input_path, wm_shape = 48, mode='bit')

        diff = [wm_extract[i] != bool_key[i] for i in range(len(wm_extract))]
        bit_acc = 1 - sum(diff) / len(diff)
        All_bit_acc += bit_acc
        #print("Bit accuracy: ", bit_acc)
    print(All_bit_acc/len(image_files))
    return All_bit_acc/len(image_files)


if __name__ == "__main__":
    # 输入和输出目录
    input_dir = "/home/jie/Program/blind_watermark/output/removal"
    
    # 批量处理图像
    Bit_acc(input_dir)

    
    
```

# RivaGAN

```python
def str2msg(str):
    return [True if el=='1' else False for el in str]
key = '111010110101000001010111010011010100010000100111'[:32] # model key
bool_key = str2msg(key)
```

```python



def batch_process_images(input_dir, output_dir, length=200, pattern=r"[wd]_\d+\.png"):
    """
    批量处理图像
    
    参数:
    input_dir: 输入图像目录
    output_dir: 输出图像目录
    pattern: 文件名匹配模式
    """
    # 确保输出目录存在
    os.makedirs(output_dir, exist_ok=True)
    
    # 获取所有图像文件
    all_files = os.listdir(input_dir)
    # 筛选符合模式的文件
    pattern = re.compile(pattern)
    image_files = [f for f in all_files if pattern.match(f)][:length]
    
    print(f"找到 {len(image_files)} 个文件符合条件")
    
    # 处理每个图像
    for img_file in tqdm(image_files, desc="处理图像"):
        input_path = os.path.join(input_dir, img_file)
        output_path = os.path.join(output_dir, img_file)
        
        bgr = cv2.imread(input_path)

        encoder = WatermarkEncoder()
        encoder.loadModel()  # 添加这一行
        encoder.set_watermark('bits', key)
        bgr_encoded = encoder.encode(bgr, 'rivaGan')

        cv2.imwrite(output_path, bgr_encoded)

if __name__ == "__main__":
    # 输入和输出目录
    input_dir = "/home/jie/Program/SG_images/val/d"
    output_dir = "/home/jie/Program/blind_watermark/output/rivaGan/rivaGan"
    
    # 批量处理图像
    batch_process_images(input_dir, output_dir, pattern=r"d_\d+\.png")

    
    print("批量处理完成！")
    
```

```python
import cv2
from imwatermark import WatermarkDecoder


def Bit_acc(input_dir, pattern=r"[wd]_\d+\.png"):
    
    All_bit_acc = 0
    decoder = WatermarkDecoder('bits', 32)

    
    all_files = os.listdir(input_dir)
    # 筛选符合模式的文件
    pattern = re.compile(pattern)
    image_files = [f for f in all_files if pattern.match(f)]
    
    print(f"找到 {len(image_files)} 个文件符合条件")
    
    # 处理每个图像
    for img_file in tqdm(image_files, desc="处理图像"):
        input_path = os.path.join(input_dir, img_file)

        bgr = cv2.imread(input_path)
        watermark = decoder.decode(bgr, 'rivaGan')
        wm_extract = [True if el==1 else False for el in watermark]
        
        diff = [wm_extract[i] != bool_key[i] for i in range(len(wm_extract))]
        bit_acc = 1 - sum(diff) / len(diff)
        All_bit_acc += bit_acc
        #print("Bit accuracy: ", bit_acc)
    print(All_bit_acc/len(image_files))
    return All_bit_acc/len(image_files)


if __name__ == "__main__":
    # 输入和输出目录
    input_dir = "/home/jie/Program/blind_watermark/output/rivaGan/removal"
    
    # 批量处理图像
    Bit_acc(input_dir)

    
    
```
